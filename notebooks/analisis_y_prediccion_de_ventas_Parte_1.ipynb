# Exploración de Datos.
1. Calcula el total de ventas por categoría de producto.

```python
# Agrupamos por Categoría. Obtenemos la columna de Monto Total de cada fila, y sumamos todas las filas por cada Categoría.
total_ventas = df.groupby("Product Category")["Total Amount"].sum()

total_ventas
```

2. Calcula el promedio de ventas diarias por categoría de producto.

```python
# Obtenemos el promedio de ventas agrupado por Categoría y Fecha.
promedio_ventas_por_cada_fecha = df.groupby(["Product Category", "Date"])["Total Amount"].mean()

# Y en base a eso, agrupamos nuevamente por categoría y obtenemos el promedio diario.
promedio_ventas_diario = promedio_ventas_por_cada_fecha.groupby("Product Category").mean()

promedio_ventas_diario
```

3. Identifica las categorías de productos con mayores y menores ventas.

```python
# Utilizando los resultados anteriores, podemos obtener el minimo y máximo del total de ventas de cada categoría.
# Ordenamos el total de ventas de forma ascendiente.
total_ventas_ordenado = total_ventas.sort_values(ascending=True)

# El máximo correspondería al último valor, y el mínimo al primer valor.
max_vendido, max_producto = total_ventas_ordenado.iloc[-1], total_ventas_ordenado.index[-1]
min_vendido, min_producto = total_ventas_ordenado.iloc[0], total_ventas_ordenado.index[0]

print(f"El producto con mayores ventas fue {max_producto} con un monto total de ventas de ${max_vendido}.")
print(f"El producto con menores ventas fue {min_producto} con un monto total de ventas de ${min_vendido}.")
```

# Manipulación de Datos:
1. Filtra los datos para mostrar solo las ventas de una categoría de producto específica.

```python
# Primero obtenemos las posibles categorías a elegir.
df["Product Category"].unique()
```

```python
# De estas quiero Clothing. Filtramos los datos para obtener solo sus ventas.
df_clothing = df[df["Product Category"] == "Clothing"]

df_clothing
```

2. Realiza operaciones de suma, resta, multiplicación y división en los datos para obtener estadísticas adicionales.

```python
# Cantidad total de productos vendidos en la categoría Clothing.
cant_total_vendida = df_clothing["Quantity"].sum()
print(f"En la Categoría Clothing se vendieron en total {cant_total_vendida} productos.")
```

```python
# La diferencia en precio entre la venta más alta de Clothing y la más baja.
max_venta = df_clothing["Total Amount"].max()
min_venta = df_clothing["Total Amount"].min()
print(f"La mayor venta de Clothing fue de ${max_venta} mientras que la menor fue de ${min_venta}.")
print(f"Por tanto al diferencia entre la venta más alta y la más baja fue de ${max_venta - min_venta}.")
```

```python
# Obtener el precio promedio de los productos vendidos en clothing.
suma_total_precio_por_producto = df_clothing["Price per Unit"].sum()
cantidad_filas = df_clothing.shape[0]
print(f"En las {cantidad_filas} filas de Clothing, el precio promedio por unidad vendida fue de ${round(suma_total_precio_por_producto/cantidad_filas,2)}.")
```
