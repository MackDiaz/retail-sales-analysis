# Cargar los Datos

```python
# Carga el archivo retail_sales.csv en un DataFrame de Pandas.
path = "../data/retail_sales_dataset.csv"
df = pd.read_csv(path)
# Muestra las primeras 10 filas del DataFrame para confirmar que los datos se han cargado correctamente.
df.head(10)
```

```python
# Aplicamos una corrección a los nombres de las columnas para mayor comodidad. En minúsculas y sin espacios.
df.columns = df.columns.str.lower().str.replace(" ","_")
```

# Filtrado de Datos

```python
# Filtra el DataFrame para mostrar solo las filas donde las ventas (Ventas) sean mayores a 50.
# La columna Total Amount viene siendo el equivalente a la venta.
df[df["total_amount"] > 50]
```

```python
# Filtra el DataFrame para mostrar solo las filas donde el precio (Precio) sea menor a 0.5.
# Lo más cercano sería el precio unitario del producto. Pero estos precios unitarios van desde $25 hasta $500.
# Por lo que modificaré el ejercicio a menor a 50.
df[df["price_per_unit"] < 50]
```

```python
# Utilizando el método query(), filtra el DataFrame para mostrar las filas donde el producto sea Manzanas y las ventas sean mayores a 30.
# Mismo caso aqui, se modifica para mostrar las ventas donde la Categoría del producto sea Beauty, y la venta sea mayor a 30.
df.query("product_category == 'Beauty' & total_amount > 30")
```

# Slicing de Datos

```python
# Selecciona y muestra solo las columnas Producto y Ventas del DataFrame.
# Se adapta a Categoría de Producto y Monto Total.
df[["product_category","total_amount"]]
```

```python
# Utilizando loc[], selecciona y muestra las filas de la 5 a la 10 (inclusive) y las columnas Producto y Tienda.
# Nuevamente, la columna Tienda no existe por lo que se adapta para el ejercicio a la columna Gender.
df.loc[5:10, ["gender", "product_category"]]
```
